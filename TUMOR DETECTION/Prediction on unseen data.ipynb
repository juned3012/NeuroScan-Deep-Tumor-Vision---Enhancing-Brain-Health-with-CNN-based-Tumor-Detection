```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
from PIL import Image
```

```python
loaded_model = tf.keras.models.load_model('CNN_Model')
```

```python
loaded_image = Image.open(R"C:\\Users\\ashwin\\Downloads\\Brain_Tumor\\Training\\meningioma_tumor\\m3 (58).jpg")
plt.imshow(loaded_image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from PIL import Image

# Load and preprocess a new image for prediction
new_image_path =R"C:\\Users\\ashwin\\Downloads\\Brain_Tumor\\Training\\meningioma_tumor\\m3 (58).jpg"
new_image = Image.open(new_image_path).convert("L")  # Convert to grayscale
new_image = new_image.resize((256, 256))  # Resize to match model's input shape
new_image_array = img_to_array(new_image)

# Normalize pixel values to [0, 1]
new_image_array = new_image_array / 255.0

# Expand dimensions to match the model's input shape
new_image_array = np.expand_dims(new_image_array, axis=0)

# Make predictions using the loaded model
predictions = loaded_model.predict(new_image_array)
predicted_class = np.argmax(predictions)

```

```python
# Display the image
plt.imshow(new_image, cmap='gray')  # Display the grayscale image
plt.title("Input Image")
plt.axis('off')
plt.show()
class_names = ['glioma_tumor','meningioma_tumor','no_tumor','pituitary_tumor']
print("Predicted class:", class_names[predicted_class])
```
